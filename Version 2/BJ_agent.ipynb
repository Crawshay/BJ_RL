```python
def gen_shoe(df_cards,No_decks):
    shoe = df_cards
    for n1 in range(1,int(np.floor(No_decks))):
        shoe = pd.concat([shoe,df_cards]).reset_index(drop=True)
    
    if No_decks % 1.0 != 0.0:
        extra = df_cards.sample(frac=1).reset_index(drop=True)[0:(int((No_decks % 1.0)*52))]
        shoe = pd.concat([shoe,extra]).reset_index(drop=True)
        
    shoe = shoe.sample(frac=1).reset_index(drop=True)
    return shoe
        
    
```

```python
def multi_shoe(df_cards,No_shoes,Decks_per_shoe):
    df_AC = gen_shoe(df_cards,Decks_per_shoe)
    df_AC["Shoe"] = 1
    for n1 in range(1,int(No_shoes)):
        df_new = gen_shoe(df_cards,Decks_per_shoe)
        df_new["Shoe"] = n1 + 1             
        df_AC = pd.concat([df_AC,df_new]).reset_index(drop=True)
    return df_AC
```
